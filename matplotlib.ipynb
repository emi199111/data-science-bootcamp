```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
```

```python
fig.show()
```

Wykres liniowy

```python
a=plt.plot([0,1],[0,3])
```

```python
a=plt.plot([0,1],[0,3], label='line', color='red')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
a=plt.legend()
```

```python
a=plt.plot([0,1,2],[0,3, 4], label='line', color='red')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
a=plt.legend()
```

```python
x = np.arange(-3, 3,0.1)
x_sin=np.sin(x)
x_cos=np.cos(x)

plt.plot(x, x_sin, label='sin(x)', color='purple')
plt.plot(x, x_cos, label='cos(x)', color='orange')

plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
a=plt.legend()
```

tworzenie wielu wykresow

```python
plt.subplot(121)
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[3,2,6])
```

```python
fig=plt.figure(figsize=(7,7))

plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,2,6], 'r--')
plt.subplot(133)
plt.plot([0,1,2],[3,2,6], 'go')
```
